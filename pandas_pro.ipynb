**Задание 1**
Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:
- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений



```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv',sep=';')
log
```

```python
def source_t(source,region):
    if source in  ['yandex','google']:
        source_type='organic'
    elif region=='Russia' and source in ['paid','email']:
        source_type='ad'
    elif region!='Russia' and source in ['paid','email']:
        source_type='other'
    else:
        source_type=source
    return source_type
```

```python
log['source_type']=log.apply(lambda x: source_t(x['traffic_source'], x['region']), axis=1)
log
```

```python
log.groupby(['source_type','region','traffic_source']).count()
```

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
data
```

```python
import pymystem3
```

```python
from pymystem3 import Mystem
```

```python
m=Mystem()

keyword = str(data['keyword'].tolist()).strip('[]')

lemmas=m.lemmatize(keyword)
```
