```python
#перечень всех документов

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

#перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

# Пользователь по команде "fio" может узнать владельца документа по его номеру

def fio( ):  
        k=input('Введите число')
        for i in documents:
            if i['number']==k:
                fio=i['name']
                break
            else:
                fio='не найден'  
        print(fio)   
        
#Пользователь по команде "p" может по номеру документа узнать на какой полке он хранится

def p( ):
    k=input('Введите номер документа')
    for key,values in directories.items():
         for i in values:
            if i==k:
                polka=key
                break
            else:
                polka='не найден'
    print(polka)
    
# Пользователь по команде "l" может увидеть полную информацию по всем документам

def l():
    for i in documents:
        tp=i['type']
        nm=i['number']
        nam=i['name']
        for key,values in directories.items():
            if i['number'] in values:
                x=key
        print('тип:',tp, 'номер:', nm, 'имя:', nam,'полка хранения:',x)
        
#Пользователь по команде "ads" может добавить новую полку

         
def ads(): 
 
    shelf_num = input('Введите номер полки: ')


    if shelf_num not in directories.keys():
        print('нет введенной полки, есть полки:')
        for key in directories:
            print(key)
        directories[shelf_num ]=[]
        print(directories)
    else:
        print('Полка уже есть')
        return 
    
#Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)

def ds():
    user_input2=str(input('введите номер полки '))
    text='полка не пустая'
    for key,values in directories.items():
        if key==user_input2 and  values==[ ]:
            text='полка удалена'
            del(directories[user_input2])
            
            break
        
    print(text)
    print('Новый перечень полок:')
    for key in directories:
            print(key)
    
    
#ввод команд

while True:
        c=str(input('введите команду '))
        if c=='узнать владельца документа по его номеру':
            print(fio())
        elif c=='по номеру документа узнать на какой полке он хранится':
            print(p())   
        elif c=='увидеть полную информацию по всем документам':
            print(l())  
        elif c=='добавить новую полку':
            print(ads()) 
        elif c=='удалить существующую полку из данных':
            print(ds()) 
            break
        else: 
            print('такой команды не существует')
              
```
